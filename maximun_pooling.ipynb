```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from matplotlib import gridspec
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing import image_dataset_from_directory
import learntools.computer_vision.visiontools as visiontools

plt.rc('figure', autolayout=True)
plt.rc('axes', labelweight='bold', labelsize='large',
       titleweight='bold', titlesize=18, titlepad=10)
plt.rc('image', cmap='magma')
```

```python
pretrained_base = tf.keras.models.load_model(
    '../input/cv-course-models/cv-course-models/vgg16-pretrained-base',
)
```

```python
model = keras.Sequential([
    pretrained_base,
    layers.GlobalAvgPool2D(),
])
```

```python
ds = image_dataset_from_directory(
    '../input/car-or-truck/train',
    labels='inferred',
    label_mode='binary',
    image_size=[128, 128],
    interpolation='nearest',
    batch_size=1,
    shuffle=True,
)
```

```python
ds_iter = iter(ds)
```

```python
sample = next(ds_iter)
```

```python
image_tf = tf.image.resize(sample[0], size=[128, 128])
```

```python
features = model(image_tf)
```

```python
features_2d = tf.reshape(features, shape=(16, 32))
```

```python
label = int(tf.squeeze(sample[1]).numpy())
class_name = ["Car", "Truck"][label]
```

```python
plt.figure(figsize=(8, 4))

plt.subplot(121)
plt.imshow(tf.squeeze(sample[0]))
plt.axis('off')
plt.title(class_name)

plt.subplot(122)
plt.imshow(features_2d)
plt.axis('off')
plt.title('Pooled Feature Maps')

plt.show()
```
